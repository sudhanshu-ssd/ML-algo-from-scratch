```python
import numpy as np
```

```python
class PCA:
    def __init__(self,n_components):
        self.n_components=n_components
        self.principal_components=None
        

    def fit(self,X):
        C=X-np.mean(X,axis=0)
        sigma=np.cov(C,rowvar=False)
        # or 
        # sigma=C.T@C/X.shape[0]
        eigenvalues,eigenvec=np.linalg.eigh(sigma)

        sorted_idx=np.argsort(eigenvalues)[::-1] #-->will give us index array of eigenvalue's array in descendinng order([::-1]) 

        sorted_values=eigenvalues[sorted_idx]
        sorted_vec=eigenvec[:,sorted_idx]
        self.principal_components=sorted_vec[:,0:self.n_components]

        # or
        # u,s,vh=np.linalg.svd(C)
        # v=vh.T
        # self.principal_components=v[:,0:self.n_components]
        

    def transform(self,X):
        X_centered=X-np.mean(X,axis=0)
        return X_centered@self.principal_components

        
        
```
