```python
import numpy as np
```

```python
class logisticregression:
    def  __init__(self,niters=1000,lr=0.01,l2_lambda=0.0):
        self.niters=niters
        self.w=None
        self.b=None
        self.lr=lr
        self.l2_lambda=l2_lambda
        self.loss_history=[]


    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
        
    
    def fit(self,X,Y):
        nsamples,nfeatures=X.shape
        self.w=np.zeros((nfeatures,1))
        self.b=0

        Y=Y.reshape(-1,1)
        
        for _ in range(self.niters):

            z=X@self.w + self.b

            y_pred=self.sigmoid(z)

            costf = -np.sum(Y * np.log(y_pred + 1e-15) + (1 - Y) * np.log(1 - y_pred + 1e-15)) + self.l2_lambda/(2*nsamples)*np.sum(self.w**2)
            self.loss_history.append(costf)

            dw=X.T@(y_pred-Y) + self.l2_lambda/nsamples*self.w
            db=np.sum(y_pred-Y)

            self.w-=self.lr*dw
            self.b-=self.lr*db


    def predict(self,X):
        y_pred = self.sigmoid(X @ self.w + self.b)
        return np.where(y_pred >= 0.5, 1, 0)

    def score(self, X, Y):
        y_pred = self.predict(X)
        return np.mean(y_pred == Y)


            

```
